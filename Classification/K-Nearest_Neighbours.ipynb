### K Nearest Neighbors (KNN)
Steps:
1. Import from Sklearn
2. Train with the data set
3. Predict outcome

For example, lets use the IRIS dataset.

```python
# importing libraries 
from sklearn import datasets
from sklearn.neighbors import KNeighborsClassifier

# Getting the data
flowers_data = datasets.load_iris()

# Printing the dataset
# flowers_data.DESCR

# Creating model
model = KNeighborsClassifier()

# Training the model
model.fit(flowers_data.data, flowers_data.target)

# Predicting the outcome
to_predict = [[1.6, 0.1, 0.1, 9.1], [0.6, 1.6, 1.1, 8.1], [0.9, 1.0, 0.1, 3.7]]
result = model.predict(to_predict)

# Defining predictions
predictions = {
    0: 'Setosa',
    1: 'Versicolour',
    2: 'Virginica'
}

# Printing the prediction
for i in range(3):
    print(predictions[result[i]])
```

### Plotting the model

```python
import matplotlib.pyplot as plt
%matplotlib inline

# Extracting the features from the flowers data
X = flowers_data.data

# Extracting the target labels
y = flowers_data.target

# Plotting the scatter plot
plt.figure(figsize=(14, 10))

# Plotting the data points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Scatter Plot of Iris Dataset')
plt.colorbar(label='Target Class')
```

#### Note:
- `c=y` specifies the color of each point based on the array `y`. Each unique value in `y` will correspond to a different color in the plot.
- `cmap='viridis'` specifies the colormap to use for mapping the values in `y` to colors. In this case, the `viridis` colormap will be used, which ranges from dark blue to yellow. Different colormaps provide different visualizations of the data.
