### Gradient Descent
An iterative optimization algorithm used to find the minimum of a function. In the context of linear regression, it can be used to minimize the cost function associated with the model parameters (e.g., MSE). It iteratively updates the parameters in the direction of the steepest descent of the cost function until convergence is reached.

```python
# Import necessary modules
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

### Steps
1. Take the data set (here arrays).
2. Define Gradient Descent Function.
    - Calculate MSE and then iterate to lower down its value to lowest.
3. Can also plot the graph for visualization.

### Formulae
#### 1. MSE
MSE = $\frac{1}{n} \Sigma_{i=1}^n({y}-\hat{y})^2$

i.e. $\frac{1}{n} \Sigma_{i=1}^n({actual}-\hat{predicted})^2$

#### 2. Partial Derivative w.r.t. m (slope)

$\frac{\partial MSE}{\partial m} = -\dfrac{2}{n} \sum\limits_{i=1}^{n} x_i (y_i - (m x_i + b))$

#### 3. Partial Derivative w.r.t. b (intercept)

$\frac{\partial MSE}{\partial b} = -\dfrac{2}{n} \sum\limits_{i=1}^{n} (y_i - (m x_i + b))$

```python
# Defining gradient descent function
def gradient_descent(x, y, learning_rate, iterations):
    # Assuming 'm' and 'b' for calculation of 'y-predicted'
    m_current = b_current = 0
    # Calculating 'n' for the further MSE calculating, 'n' will be the length of the data points
    n = len(x)

    for i in range(iterations):
        # y = m*x + b
        y_predicted = m_current * x + b_current 
        # Calculating cost or MSE
        cost = (1/n) * sum([value**2 for value in (y - y_predicted)])

        # Calculating partial derivative of 'm' and 'b'
        derivative_m = -(2/n) * sum(x * (y - y_predicted))
        derivative_b = -(2/n) * sum((y - y_predicted))

        # Updating the 'm_current' and 'b_current'
        m_current = m_current - learning_rate * derivative_m
        b_current = b_current - learning_rate * derivative_b

        # Printing values for clarity
        print(f'm = {m_current}, b = {b_current}, iteration = {i}, MSE = {cost}')



# Assume some random value of learning_rate and iteration to start with
learning_rate = 0.08
iterations = 1000
# Calling the function
gradient_descent(x, y, learning_rate, iterations )
```

We can also plot the scatter plot and check for each iteration and see how best fit line is getting created. So, before running the each iteration, we will print the current line.


```python
def gradient_descent(x, y, learning_rate, iterations):
    m_current = b_current = 0
    n = len(x)

    # PLOTTING the x and y values
    plt.figure(figsize=(14, 10)) 
    plt.scatter(x, y, color='red', marker='+', linewidths=2, s=100)

    for i in range(iterations):
        y_predicted = m_current * x + b_current 

        # PLOTTING the current best fit line
        plt.plot(x, y_predicted, color='yellow')

        derivative_m = -(2/n) * sum(x * (y - y_predicted))
        derivative_b = -(2/n) * sum((y - y_predicted))
        m_current = m_current - learning_rate * derivative_m
        b_current = b_current - learning_rate * derivative_b

    # Plot the best fit line with a different color
    plt.plot(x, y_predicted, color='blue', label='Best Fit Line')
    plt.legend()  # Show legend

learning_rate = 0.08
iterations = 1000
gradient_descent(x, y, learning_rate, iterations )
```
