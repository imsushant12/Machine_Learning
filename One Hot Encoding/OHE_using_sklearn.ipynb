```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
import pandas as pd
```

```python
# Loading data
data = pd.read_csv('homedata.csv')

# Creating object of label encoder
labelEncoder = LabelEncoder()

# Encoding/ Transforming the categorical data (town) into numerical data
data.town = labelEncoder.fit_transform(data.town)

data
```

```python
# Extracting X and Y values for training the model
X = data[['town', 'area']]
Y = data[['price']]
```

```python
# Training the model
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X, Y)
```

```python
# Predicting outcomes
print(model.predict([[0,3400]])) # 3400 sqr ft home in west windsor
print(model.predict([[1,2800]])) # 2800 sqr ft home in robbinsville
print(model.predict([[2,4800]])) # 2800 sqr ft home in monroe township
```

```python
# Plotting the graph
plt.plot(X, Y)
```
