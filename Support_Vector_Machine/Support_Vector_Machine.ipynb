```python
# Importing Necessary Modules
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris
%matplotlib inline
```

```python
# Loading dataset
iris = load_iris()

iris
```

```python
# Convert dataset into a DataFrame for data visualization and easy working
data = pd.DataFrame(iris.data, columns=iris.feature_names)
data
```

Adding a target column, having values:
- 0 means setosa
- 1 means versicolor
- 2 means virginica

We can also add another column (`flower_name`) that will show the name of the flower. For example it will show setosa where the target value is 0. We can use the `apply()` function and a lambda function on the `iris.target_names` as it contains the name of the flowers.

```python
# Adding target column
data['target'] = iris.target

# Adding flower_names column
data['flower_name'] = data.target.apply(lambda x: iris.target_names[x])

data
```

We can see that the data sets are divided into 3 categories:
- 0 to 49 index: sentosa
- 50 to 99 index: versicolor
- 100 to 149 index: virginica

So, we can create 3 separate dataframe for testing and visualization.

```python
df_0 = data[:50]
df_1 = data[50:100]
df_2 = data[100:]
```

Lets now visualize: Sepal length vs Sepal Width for various combinations

```python
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.scatter(df_0['sepal length (cm)'], df_0['sepal width (cm)'],color="orange",marker='*', label='Sentosa')
plt.scatter(df_1['sepal length (cm)'], df_1['sepal width (cm)'],color="red",marker='.', label='Versicolor')
plt.scatter(df_2['sepal length (cm)'], df_2['sepal width (cm)'],color="turquoise",marker='+', label='Virginica')
plt.legend()
plt.figure(figsize=(12, 6))
plt.show()
```

As we can see that we can easily draw line to differentiate between sentosa and versicolor and same for sentosa and virginica

We can similarly draw graph for Petal Length vs Petal Width

```python
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.scatter(df_0['petal length (cm)'], df_0['petal width (cm)'],color="orange",marker='*', label='Sentosa')
plt.scatter(df_1['petal length (cm)'], df_1['petal width (cm)'],color="red",marker='.', label='Versicolor')
plt.scatter(df_2['petal length (cm)'], df_2['petal width (cm)'],color="turquoise",marker='+', label='Virginica')
plt.legend()
plt.figure(figsize=(12, 6))
plt.show()
```

Let's train our model but first let's divide our train and test data

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

X = data.drop(['target','flower_name'], axis='columns')
Y = data.target

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)

# Creating object of Support Vector Machine
model = SVC()

# Training model
model.fit(X_train, Y_train)

# Checking score of model
model.score(X_test, Y_test)
```
